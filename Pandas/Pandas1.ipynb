Lable based Indexing

```python
people_age = {
    "Rahul": 21,
    "Anita": 19,
    "Suresh": 25,
    "Priya": 22,
    "Amit": 24,
    "Neha": 20
}
series = pd.Series(people_age,name='Age')
series
```

CONDITIONAL INDEXING

```python
print(series>20)
series[series>20]
```

Logical Operation

```python
series[(series>20) & (series<30)]
```

```python
series['Rahul']=22 # modifying the colomn
series
```
