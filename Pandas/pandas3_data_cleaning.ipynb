```python
import pandas as pd
```

```python
import numpy as np
```

Data Frames


```python
data = {
    "names":["pranav","kerkar", "rahul", "rohit"],
    "age":[12,13,np.nan,19],
    "dept":["it","comp","mech","ecs"],
    "salary":[12000,13000,np.nan,20000]
}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
print(df.head(3))
print(df.tail(1))
```

```python
df.iloc[0:3] # only the last value is not considered
```

```python
df.loc[0:3] # all index values are considered
```

```python
print(df["age"]) # to access single col
print(df[["age",'dept']]) # two brackets for accessing two colouns
```

```python
# to drop the coloumn
df.drop("age",axis=1,inplace=True) # axis 1 means full coloum and axis 0 means first row, inplace tells to drop the col in the original table also.
```

```python
print(df)
```

```python
print(df.shape)
print(df.info())
```

Broadcasting

```python
df['salary']=df['salary'] + 5000
df['salary']
```

Renaming coloumn

```python
df.rename(columns = {'dept':"Department"},inplace = True)
print(df)
```

```python
print(df['Department'].unique())
print(df['Department'].value_counts())
```
