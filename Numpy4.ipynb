Vectorizarion


```python
import numpy as np
arr = np.array([100,2,3,4])
def square(x):
  x*x

vfucn = np.vectorize(square)
print(arr)
print(vfucn(arr))
```

Dealing with Missing Values

```python
# np.nan
arr = np.array([1,np.nan,2,3])
print(arr)
print(np.isnan(arr))
# np.isinf, -np.isinfn, np.isfinite
```

Replacing the Missing Values

```python
print(np.nan_to_num(arr)) # replaces with 0s
```
